# Tuples

tuples are immutable

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.  

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.  

Characteristics
Ordered

Unchangeable

Allows duplicate

# Creating a tuple

```python
# empty tuple
t1 = ()
print(t1)

# single tuple
t1 = ('abhinav',)
print(t1)

# heterogeneous
t1 = ('abhinav', 12,23.3, True)
print(t1)

 # homogeneous tupl3
t1 = (1,2,3,4,5)
print(t1)

# type conversion
t1 = 'harivansh'
type(tuple(t1))

# 2D tuple
t1  = ((2,3),(5,6))
print(t1)
```

Sets
A set is an unordered collection of items. Every set element [is unique (has no duplicates) and must be immutable (cannot be changed)].

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection,symmetric difference

Characteristics:

Unordered

Mutable

No Duplicates

Can't contain mutable data types

```python
#  set is unhashable coz it is mutable
```

#Creating a set

```python
# creating a empty set
# so default is dictionary we use set conversion to create empty list

s = {}
print(type(s))
print(type(set(s)))

# # creating  a 2d set
# s = {{1,2,3},{4,5}} # 2d sets are not allowed
# print(s)

# s = {1,2,3,4,54}
# print(s)


# s= {1,2,3,4.4,False}
# print(s)

 # duplicates are not allowed
s = {1,1,2,3,4,4}
print(s)

# # can't store mutable items in a set
# s = {1,2,3,4,[2,3]}
# print(s)

#immutable items are allowed
s = {1,2,3,4,(2,3)}
print(s)



# order doesn't matter
s = {1.2,3}
s1 ={3,1.2}
print(s == s1)


# type conversion
L = [1,23,3]
print(type(set(L)))
```

# set operations

```python
# union
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 | s2 # it will give all unique values
```

```python
# intersection

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1 & s2
```

```python
# difference

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 - s2 # the items of s1 that are not present in s2
s2 - s1 # the items of s2 that are not present in s1
```

```python
# symmetric difference = everything will be printed except common

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1 ^ s2
```

```python
# membership test

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

1 in s1
```

```python
# iteration

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

for i in s1:
  print(i,end = ' ')
```

#Set functions

```python
# len , max, min, sorted, sum

s1 = {1,2,3,5,4}
len(s1)
max(s1)
min(s1)
sorted(s1)
sum(s1)
```

```python
# union/update

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.union(s2)

s1.update(s2) # it makes the permanent changes in left side set that is s1 here
s1
print(s2)
```

```python
# intersection/ intersection update

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)
s1.intersection_update(s2) # permanent changes
s1
s2
```

```python
# difference
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference()
s1.difference(s2)
s1.difference_update(s2)
s1
```

```python
# symmetric difference

s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2)
s1
s2
```

```python
# isDisjoint/ isSubset/ isSuperset

# isDisjoint = the sets which has not a single common item
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.isdisjoint(s2)
```

```python
# isSubset

s1 = {1,2,3,4,5}
s2 = {2,3,4}

s2.issubset(s1)
```

```python
# is superset
s1.issuperset(s2)
```

```python
# copy
s1 = {1,2,3,4,5}
s2 = s1.copy()
print(id(s1))
print(id(s2))
```

# Dictionary comprehension

key : value for var in iterable

```python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
# distances
distances = {'delhi': 1000, 'mumbai':2000,'banglore':3000}
print(distances.items())
```

```python
{key:value*0.62 for (key,value) in distances.items()}
```

```python
# using Zip
days = ['sunday','monday','tuesday','wednesday','thursday','friday','saturday','sunday']
temp = [23.3, 23.4,23.3, 5.54,32.4,98.8,76.8]
{i:j for (i,j) in zip(days,temp)}
```

```python
# using if conditions
products = {'phone': 23, 'laptop':34, 'charger':0, 'tablet': 23}

{key:value for (key,value) in products.items() if value != 0}
```

```python
# Nested comprehension
# print table form number 2

{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
