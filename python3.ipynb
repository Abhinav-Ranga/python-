```python
# palindrome

s = input("enter the string:")
c = ''

for i in s[-1::-1]:
  c = c+i

if s == c :
  print('this is a palindrome')
else :
  print('this is not a palindrome')
```

```python
s = input('Enter the string: ')

flag = True
for i in range(0,len(s)//2):
  if s[i] != s[len(s)-i-1]:
    flag = False
    break
if flag == True :
  print('this is palindrome')
else :
  print("this is not palindrome")
```

```python
# write a program to count the number of words in a string without using the split functions

s = input("Enter the string: ")
temp = ''
L = []


for i in s :
  if i != ' ':
    temp += i
  else :
    L.append(temp)
    temp = ''

L.append(temp)
print(L)
```

```python
# write a program to convert the string into title case without using the title () funtion
s = input("Enter the string: ")
L = []
for i in s.split():
  L.append(i[0].upper()+i[1:].lower())

print(" ".join(L))
```

```python
# write a program that convert integer to string without using str function

number = int(input("Enter the number: "))

digits = '0123456789'
result = ''

while number !=0 :
  result = digits[number%10]+ result
  number = number // 10


print(type(result))
```
