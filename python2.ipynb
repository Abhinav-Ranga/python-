```python
# Guessing random

# generate a random number between 1 to 100

import random
jackpot = random.randint(1,100)
guess = int(input("Guess the number: "))
counter = 1

while guess != jackpot :

  if guess<jackpot :
    print("guess is lower")
  else :
    print(" guess is higher")

  guess = int(input("Guess the number:"))
  counter += 1

else :
  print("guess is correct")
  print("total attempts : ",counter, guess)
```
