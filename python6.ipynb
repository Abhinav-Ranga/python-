```python
# default argument
# in default argumnent  u write the default values if user don't write those values or write
# lesser than actual parameters so will not get any error

def power(a=1,b=1):
  return a ** b

power(2)
```

# * args and   ** kwargs

 are special python keywords that are used to pass the variable length of arguments to function

```python
#  *args
# if u want to make a function that can receive multiple inputs from the user for that we
# use args
# *args allows us to pass variable number of non keyword arguments to a function


def multi(*args): # we can use any name instead of args like *radha
  product = 1
  for i in args:
    product *= i
  print(args)# it is storing the values in tuples
  print(product)

multi(2,3,3,34,23,23,23,45,3454,56564,56,345,345,34,23,2,31,321,21,2342,4,324,2354)
```

```python
# ** kwargs
# **kwargs allows us to pass variable number of  keyword arguments to a function
# keyword argument means we can pass a key value pair like python dictionary


def display(** radha) :
  for (key,value) in radha.items():
    print(key,'->',value)


display(india = 'delhi',pakistan = 'murdabad', america = 'usa') # we are not using the keys in string form like this 'india' coz we are sending through keyword not dictionary
```

# Lambda Function

A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```python
# x = x^2
a = lambda x:x**2
a(3)
```

```python
# x,y -> x+y
a = lambda x,y: x+y
a(1,2)
```

# Higher order functions

```python
# now here we are calculating square but what if we had to calculate cube or any other then
# we would need to create another function but by using lambda we can do it in one line easily

def square(x):  # without lambda
  return x**2

def transformer(f,L):
  output  = []

  for i in L:
    output.append(f(i))

  return output

L = [1,2,3,4,5]
# transformer(square,L) without lambda

transformer(lambda x:x**3,L)#  with lambda
```

#Map

with map u can apply any logic on a list

```python
# square the items of the list

list(map(lambda x:x**2 , [1,2,3,4,5]))
```

```python
# odd even labeling of the list items
list(map(lambda x: 'even' if x%2 == 0 else 'odd',[1,2,3,4]))
```

#Filter

```python
# number greater than 5
L = [1,2,3,55,6]
list(filter(lambda x:x>5,L))
```

```python
# fetch fruits starting with a
F = ['apple','mango','avacado']
list(filter(lambda F:F.startswith('a'),F))
```
