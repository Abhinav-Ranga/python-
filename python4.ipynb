# Editing items in a list

```python
# Editing with indexing
L = [1,2,3,4,5]
L[-5] = 99

# Editing with slicing
L[3:5] = [324,234]

L[-1:-4:-1] = [12,24,354]
print(L)
```

# Deleting items from a list

```python
L = [1,2,3,4,5]
# print(L)
# del L # = delete whole list
del L[2] # delete by using index
del L[0:3]
print(L)
```

```python
# remove = it deletes the value directly without taking the index
L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
# pop = it removes by taking index and at default removes the last value
L = [1,2,3,4,5]
L.pop(0)
L.pop()
print(L)
```

```python
# clear
L = [1,2,3,4,5]
L.clear()
print(L)
```

# List Comprehension
List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]

# Advantages of List Comprehension
  
More time-efficient and space-efficient than loops.  

Require fewer lines of code.  

Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers in a list

# L = []
# for i in range(1,11) :
#     L.append(i)
# print(L)

L  = [i for i in range(1,11)]
print(L)

```

```python
# scalar multiplication on vector

s = [2,3,4]
v = 2

# X  = []

# for i in s :
#   X.append(i*v)

# print(X)

L = [v*i for i in s]
print(L)



```

```python
# Add squares

# s = [2,4,6]
# l = []
# for i in s:
#   l.append(i*i)
# print(l)


l = [2,4,6]
[i**2 for i in l]
```

```python
# print all numbers divisible by 5 in range 1 to 50

# for i in range(1,51):
#   if i%5 == 0 :
#     print(i)

[i for i in range(1,51) if i%5 ==0]
```

```python
# find the languages which start with letter p

# L = ['java', 'php', 'python']

# for i in L :
#   if(i.startswith('p')):
#     print(i)

[i for i in L if i.startswith('p')]
```

```python
# nested if with list comprehnsion

# add new list from my_fruits if the item exists in basket also and starts with a

basket = ['apple', 'guvava','banana']
my_fruits = ['apple','kiwi','mango']

# [fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]

for i in my_fruits:
  if i in basket:
    if i.startswith('a'):
      print(i)
```

```python
# print a 3 *3 matrix using list comprehension -> nested list comprehension

# matrix = []

# for i in range(0,3):
#   row = []
#   for j in range(0,3):
#     row.append(j )
#   matrix.append(row)
# print(matrix)

[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> list comprehension on 2 lists together
l1 = [1,2,3,4]
l2 = [5,6,7,8]

# [i*j for i in l1 for  j in l2]

result = []


for i in l1:
  for j in l2:
    result.append(i*j)
print(result)
```
